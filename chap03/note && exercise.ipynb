```python
# 평년과 윤년 구분하는 프로그램
# 년도가 4로 나뉘어지고 100으로 나뉘어지지 않거나
# 400으로 나뉘어지면 윤년이다 

year = int(input("Type a year :"))
if (year%4 == 0 and year%100 != 0) or (year%400) == 0 :
    print(year, "is a leap year")
else :
    print(year, "is not a leap year")
```

# if-elif-else 문

# while 문
- 조건이 만족되면 블록을 실행하고 다시 조건을 조사
- 조건이 만족되지 않으면 다음 문장 실행

```python
i = 1  # 초기화
sum = 0
while i <= 100:  # 조건 판단. 끝에 :를 붙인다
    sum = sum + i  # 블록
    i = i + 1      # 조건 변경 (반복문 안에 있어야 함)
print("sum=", sum)  # 블록이 끝나면 다시 들여쓰기를 원래 위치로
```

# while 연습문제
- ❖ 10진수를 입력 받아 2진수로 변환하여 출력하는 프로그램

```python
n = int(input('Number: '))
result = ''  # 변환 결과 저장
# 10진수를 2로 나누어 몫이 0이 될 때까지 반복해서 나머지를 구한다
while n != 0:
    m = n % 2  # 나머지
    result = str(m) + result  # 숫자를 문자열로 변환
    n = n // 2  # 몫
print("2진수 =", result)
```

# for 제어문
- for 문은 일정 횟수만큼 블록을 반복실행하는 제어문이다

## for문
- [반복객체]의 각 요소의 값을 하나씩 [변수]에 저장하고 블록을 실행
- [반복객체]에 더 이상의 요소가 없을 때까지 반복
- 튜플, 문자열과 같은 반복 객체를 사용할 수도 있다.

# for-range 문
- range() 함수는 반복 객체 (iterable)를 생성하는 함수
- range(start, stop, step)
- 예) list(range(1,5,1) → [1,2,3,4]
- 변수 값을 start부터 시작하여 step 만큼 증가시키며 stop될 때까지 실행. step=1이면 생략

```python
for i in range(1, 4, 1) : #1부터 시작하여 1씩 증가하여 4가 되면 종료
    print("Number", i)
```

```python
# for 문을 이용한 1부터 100까지의 합 구하기

sum = 0
for i in range(1, 101, 1) :
    sum = sum + i
print("sum = ", sum)
```

```python
#x^n을 구하는 프로그램

x = float(input('Type x : '))
n = int(input('Type n : '))
prod = 1
for i in range(1, n+1):
    prod = prod * x
print(prod)
```

# 제어문 예제(반복문 안의 조건문)

- 숫자 맞추기 게임 프로그램
- 1~100 사이의 난수를 만들고 사용자가 1~100 사이의 수를 입력하여 맞추는 프로그램. 5회까지
맞추지 못하면 종료

```python
from random import randint
secret_num = randint(1,100)
num_guesses = 0 #시도 회수
guess = 0 #예상 숫자
while guess != secret_num and num_guesses <= 4:
    guess = eval(input('Enter your guess (1-100): ')) #수 입력
num_guesses = num_guesses + 1 #시도 횟수 증가
if guess < secret_num: #정답이 클 때
    print('더 큽니다.', 5-num_guesses, '회 남았습니다.\n')
elif guess > secret_num: #정답이 작을 때
    print('더 작습니다.', 5-num_guesses, '회 남았습니다.\n')
else: #정답
    print('맞았습니다!')
if num_guesses==5 and guess != secret_num: #맞추지 못한 경우
    print('당신이 졌습니다. 정답은 ', secret_num, '입니다')
```

## continue 문
- 블록의 나머지 부분은 생략하고 다시 while 조건 조사

```python
n = 10
while n >= -10:
    if n == 0:
        break  # n=0이면 while 블록을 빠져나온다
    inv = 1.0 / n
    print(inv)
    n = n - 1
```

```python
n = 10
while n >= -10:
    if n == 0:
        n = n - 1
        continue  # n=0이면 while 블록의 나머지 부분을 건너 뛴다
    inv = 1.0 / n
    print("n = %d, inv(n) = %0.2f" % (n, inv))
    n = n - 1
```
