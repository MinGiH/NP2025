# 모듈이란?
- 함수, 변수, 클래스들을 모아 놓은 파일 묶음(라이브러리)

# 모듈 불러오기
- 프로그램 외부에 존재하는 모듈을 프로그램 내부로 불러와서 사용하기 위해 필요한 명령
- (import)
- ① import module_name
-  '>>>' import socket
- • socket 모듈에 포함된 변수, 함수, 클래스를 사용할 수 있다
- • dir(socket) socket 모듈에 들어 있는 변수, 함수, 클래스를 보여줌
- • dir(socket.socket) socket 모듈의 socket 클래스에 포함된 변수, 메소드를 보여 줌

- ② from module_name import class or methods
- • from socket import gethostname
- • import된 class 또는 methods 사용할 수 있다

- ③ from module_name import *
- • from socket import *
- • socket 모듈의 모든 속성을 사용할 수 있다

- 주의! 편리하지만 메모리를 많이 차지하고 다른 함수와 혼동 가능성이 있다

- ④ import module_name as alias(단축이름)
- • 모듈 이름이 긴 경우 단축이름으로 불러오기

- • 수학적 계산을 위한 모듈

- • 시간을 문자열로 표시하는 모듈

- calendar 모듈

# __init__.py의 역할
- • 파이썬 패키지로 인식:
- __init__.py 파일은 해당 디렉토리를 Python 패키지로 인식하게 만듭니다.
- Python 3.3 이상에서는 __init__.py가 없어도 패키지로 인식되지만, 명시적으로 포함하는 것이 권장됩니다.
- • 초기화 코드 실행:
- 패키지가 import될 때 실행할 초기화 코드를 포함할 수 있습니다.
- 예를 들어, 패키지 내에서 공통적으로 사용할 설정이나 변수 초기화를 수행할 수 있습니다.
- • __init__.py의 위치
- mypackage/
- │
- ├── __init__.py # 패키지 초기화 파일
- ├── module1.py
- └── module2.py

# 모듈 (Module)
- • 정의: Python에서 하나의 .py 파일이 하나의 모듈
- 모듈은 함수, 클래스, 변수 등을 포함할 수 있으며, 코드 재사용을 위해 사용
#  패키지 (Package)
- • 정의: Python에서 디렉토리가 패키지
- 패키지는 여러 모듈을 포함할 수 있음
- 대규모 프로젝트에 필요
- 디렉토리에 __init__.py 가 필수적 (비어 있더라도 인식)

# 패키지 (Package)
- • 정의:
- Python에서 디렉토리가 패키지
- 패키지는 여러 모듈을 포함할 수 있으며, 디렉토리 내에 __init__.py 파일이 있어야 패키지로 인식됨
- • 특징:
- 디렉토리 단위로 구성됨.
- 모듈을 그룹화하여 계층 구조를 제공.
- 대규모 프로젝트에서 코드 구조를 체계적으로 관리하는 데 유용.

# 가상환경이란?
- 프로젝트별로 독립된 파이썬 환경을 만드는 도구
 - 각 프로젝트가 서로 다른 패키지 버전을 사용할 수 있게 함
 - 시스템 파이썬 환경을 오염시키지 않음
-  만드는 방법
- • 1. venv (파이썬 3.3+ 내장 모듈)
- • 2. conda (anaconda)
- • 3. virtualenv (써드 파티 도구)
